```python
from sklearn.preprocessing import StandardScaler
from sklearn.cluster import KMeans

# Standardize the two numeric variables: Age and Income
scaler = StandardScaler()
df_scaled = df.copy()
df_scaled[['Age', 'Income']] = scaler.fit_transform(df[['Age', 'Income']])
```

```python
inertias = []

for i in range(1, 11):
    km = KMeans(n_clusters=i,
                init='k-means++',
                n_init=10,
                max_iter=300,
                random_state=0)
    km.fit(df_scaled)
    inertias.append(km.inertia_)

# Plotting the Elbow Method graph
plt.figure(figsize=(10, 6))
plt.plot(range(1, 11), inertias, marker='o', linestyle='--')
plt.xlabel('Number of Clusters')
plt.ylabel('Cluster Inertia (Within-Cluster SSE)')
plt.title('Elbow Method for Optimal Number of Clusters')
plt.xticks(range(1, 11))
plt.tight_layout()
plt.savefig('/Users/ander/Downloads/Chart/elbowmethod.png')
plt.show()
```

```python
import numpy as np
from sklearn.metrics import silhouette_samples
from sklearn.cluster import KMeans

# Computing Silhouette Average Score for k = 2, 3, 4

# k = 2
km2 = KMeans(n_clusters=2,
             init='k-means++',
             n_init=10,
             max_iter=300,
             tol=1e-04,
             random_state=0)
y_km2 = km2.fit_predict(df_scaled)
silhouette_vals_2 = silhouette_samples(df_scaled, y_km2, metric='euclidean')
silhouette_avg_2 = np.mean(silhouette_vals_2)
print(f'Silhouette average for k=2: {silhouette_avg_2:.2f}')

# k = 3
km3 = KMeans(n_clusters=3,
             init='k-means++',
             n_init=10,
             max_iter=300,
             tol=1e-04,
             random_state=0)
y_km3 = km3.fit_predict(df_scaled)
silhouette_vals_3 = silhouette_samples(df_scaled, y_km3, metric='euclidean')
silhouette_avg_3 = np.mean(silhouette_vals_3)
print(f'Silhouette average for k=3: {silhouette_avg_3:.2f}')

# k = 4
km4 = KMeans(n_clusters=4,
             init='k-means++',
             n_init=10,
             max_iter=300,
             tol=1e-04,
             random_state=0)
y_km4 = km4.fit_predict(df_scaled)
silhouette_vals_4 = silhouette_samples(df_scaled, y_km4, metric='euclidean')
silhouette_avg_4 = np.mean(silhouette_vals_4)
print(f'Silhouette average for k=4: {silhouette_avg_4:.2f}')
```

```python
from matplotlib import cm
import numpy as np
from sklearn.metrics import silhouette_samples
from sklearn.cluster import KMeans
import matplotlib.pyplot as plt

# Silhouette Plot for k = 2, 3, 4
fig, axes = plt.subplots(1, 3, figsize=(18, 6))

# k = 2
km2 = KMeans(n_clusters=2, init='k-means++', n_init=10, max_iter=300, tol=1e-04, random_state=0)
y_km2 = km2.fit_predict(df_scaled)
silhouette_vals_2 = silhouette_samples(df_scaled, y_km2, metric='euclidean')
silhouette_avg_2 = np.mean(silhouette_vals_2)
cluster_labels_2 = np.unique(y_km2)
n_clusters_2 = cluster_labels_2.shape[0]
y_lower, yticks = 0, []
for i, c in enumerate(cluster_labels_2):
    c_sil_vals = silhouette_vals_2[y_km2 == c]
    c_sil_vals.sort()
    y_upper = y_lower + len(c_sil_vals)
    color = cm.jet(float(i) / n_clusters_2)
    axes[0].barh(range(y_lower, y_upper), c_sil_vals, height=1.0, edgecolor='none', color=color)
    yticks.append((y_lower + y_upper) / 2.)
    y_lower = y_upper
axes[0].axvline(silhouette_avg_2, color="red", linestyle="--")
axes[0].set_title('Silhouette Plot (k=2)')
axes[0].set_xlabel('Silhouette Coefficient')
axes[0].set_ylabel('Cluster')
axes[0].set_yticks(yticks)
axes[0].set_yticklabels(cluster_labels_2 + 1)

# k = 3
km3 = KMeans(n_clusters=3, init='k-means++', n_init=10, max_iter=300, tol=1e-04, random_state=0)
y_km3 = km3.fit_predict(df_scaled)
silhouette_vals_3 = silhouette_samples(df_scaled, y_km3, metric='euclidean')
silhouette_avg_3 = np.mean(silhouette_vals_3)
cluster_labels_3 = np.unique(y_km3)
n_clusters_3 = cluster_labels_3.shape[0]
y_lower, yticks = 0, []
for i, c in enumerate(cluster_labels_3):
    c_sil_vals = silhouette_vals_3[y_km3 == c]
    c_sil_vals.sort()
    y_upper = y_lower + len(c_sil_vals)
    color = cm.jet(float(i) / n_clusters_3)
    axes[1].barh(range(y_lower, y_upper), c_sil_vals, height=1.0, edgecolor='none', color=color)
    yticks.append((y_lower + y_upper) / 2.)
    y_lower = y_upper
axes[1].axvline(silhouette_avg_3, color="red", linestyle="--")
axes[1].set_title('Silhouette Plot (k=3)')
axes[1].set_xlabel('Silhouette Coefficient')
axes[1].set_yticks(yticks)
axes[1].set_yticklabels(cluster_labels_3 + 1)

# k = 4
km4 = KMeans(n_clusters=4, init='k-means++', n_init=10, max_iter=300, tol=1e-04, random_state=0)
y_km4 = km4.fit_predict(df_scaled)
silhouette_vals_4 = silhouette_samples(df_scaled, y_km4, metric='euclidean')
silhouette_avg_4 = np.mean(silhouette_vals_4)
cluster_labels_4 = np.unique(y_km4)
n_clusters_4 = cluster_labels_4.shape[0]
y_lower, yticks = 0, []
for i, c in enumerate(cluster_labels_4):
    c_sil_vals = silhouette_vals_4[y_km4 == c]
    c_sil_vals.sort()
    y_upper = y_lower + len(c_sil_vals)
    color = cm.jet(float(i) / n_clusters_4)
    axes[2].barh(range(y_lower, y_upper), c_sil_vals, height=1.0, edgecolor='none', color=color)
    yticks.append((y_lower + y_upper) / 2.)
    y_lower = y_upper
axes[2].axvline(silhouette_avg_4, color="red", linestyle="--")
axes[2].set_title('Silhouette Plot (k=4)')
axes[2].set_xlabel('Silhouette Coefficient')
axes[2].set_yticks(yticks)
axes[2].set_yticklabels(cluster_labels_4 + 1)

# Final layout and save
plt.tight_layout()
plt.savefig('/Users/ander/Downloads/Chart/silhouette_combined.png')
plt.show()
```

```python
# K-means++ Cluster Summary
# Apply K-means++ clustering
kmeans = KMeans(n_clusters=2, init='k-means++', random_state=42)
df['KMeans_Cluster'] = kmeans.fit_predict(df_scaled)

# Create summary table for K-means++ clusters
kmeans_summary = df.groupby('KMeans_Cluster').agg({
    'Gender': 'mean',
    'Marital Status': 'mean',
    'Age': 'mean',
    'Education': 'mean',
    'Income': 'mean',
    'Occupation': 'mean',
    'Settlement Size': 'mean',
    'KMeans_Cluster': 'count'  # Count how many members in each cluster
}).rename(columns={'KMeans_Cluster': 'Count'})

# Display summary table
kmeans_summary_reset = kmeans_summary.reset_index().round(3)
print("K-means++ Cluster Summary:")
kmeans_summary_reset
```

```python
from sklearn.cluster import AgglomerativeClustering

# Apply Agglomerative Clustering
agglo = AgglomerativeClustering(n_clusters=2)
df['Agglo_Cluster'] = agglo.fit_predict(df_scaled)

# Create summary table for Agglomerative clusters
agglo_summary = df.groupby('Agglo_Cluster').agg({
    'Gender': 'mean',
    'Marital Status': 'mean',
    'Age': 'mean',
    'Education': 'mean',
    'Income': 'mean',
    'Occupation': 'mean',
    'Settlement Size': 'mean',
    'Agglo_Cluster': 'count'  # Count how many members in each cluster
}).rename(columns={'Agglo_Cluster': 'Count'})

# Display summary table
agglo_summary_reset = agglo_summary.reset_index().round(3)
print("Agglomerative Clustering Cluster Summary:")
agglo_summary_reset
```
